```python
import yfinance as yf
import pandas as pd
```

```python
EURUSD_Exchange = yf.download("EURUSD=X", period="2y", interval='1h')
EURUSD_Exchange.iloc[:,:]
```

```python
EURUSD_Exchange.index = EURUSD_Exchange.index.tz_convert(None)
EURUSD_Exchange
```

```python
EURUSD_Exchange.info()
```

```python
EURUSD_Exchange.mean()
```

```python
EURUSD_Exchange.max()
```

```python
EURUSD_Exchange.min()
```

```python
EURUSD_Exchange['Close_Difference'] = EURUSD_Exchange['Close'].diff()
EURUSD_Exchange
```

```python
EURUSD_Exchange['Close_Percentage_Difference'] = EURUSD_Exchange['Close'].pct_change() * 100
EURUSD_Exchange
```

```python
EURUSD_Exchange['Daily_Return'] = EURUSD_Exchange['Close'].pct_change()
print(EURUSD_Exchange['Daily_Return'])
```

```python
EURUSD_Exchange['SMA_20'] = EURUSD_Exchange['Close'].rolling(window=20).mean()
print(EURUSD_Exchange['SMA_20'])
```

```python
EURUSD_Exchange['Volatility'] = EURUSD_Exchange['Daily_Return'].rolling(window=20).std()
print(EURUSD_Exchange['Volatility'])
```

```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import mpl_finance as mpf
from mpl_finance import candlestick_ohlc
```

```python
ohlc = EURUSD_Exchange[['Open', 'High', 'Low', 'Close']]
ohlc['Date'] = mdates.date2num(ohlc.index)
ohlc_values = ohlc[['Date', 'Open', 'High', 'Low', 'Close']].values

# Create a subplot with candlestick chart
fig, ax = plt.subplots(figsize=(10, 6))
mpf.candlestick_ohlc(ax, ohlc_values, width=0.6, colorup='g', colordown='r')

# Format the x-axis as dates
ax.xaxis_date()
plt.xticks(rotation=45)

# Set labels and title
plt.xlabel('Date')
plt.ylabel('Price')
plt.title('EUR/USD Candlestick Chart')

# Show the plot
plt.show()
```
